```python
### distribuição salarial para cada profissão ###

# absoluto
for prof in profissoes:
    # print(prof)
    df_prof = df[df["cargo"] == prof]
    abs_genero = pd.crosstab(df_prof["salario"],df_prof["genero"]).reset_index()

    larguras = {500: {"min": 0, "max": 1000},
    1500: {"min": 1000, "max": 2000},
    2500: {"min": 2000, "max": 3000},
    3500: {"min": 3000, "max": 4000},
    5000: {"min": 4000, "max": 6000},
    7000: {"min": 6000, "max": 8000},
    10000: {"min": 8000, "max": 12000},
    14000: {"min": 12000, "max": 16000},
    18000: {"min": 16000, "max": 20000},
    22500: {"min": 20000, "max": 25000},
    27500: {"min": 25000, "max": 30000},
    35000: {"min": 30000, "max": 40000},
    45000: {"min": 40000, "max": 50000}}

    # Usando list comprehension corretamente
    abs_genero["min"] = [larguras.get(sal, {}).get("min", None) for sal in abs_genero["salario"]]
    abs_genero["max"] = [larguras.get(sal, {}).get("max", None) for sal in abs_genero["salario"]]

    # Preenchendo os valores de 500 em 500
    salarios = pd.Series(range(0, 50001, 500), name='salario')

    # Vamos criar um DataFrame para armazenar os dados repetidos para cada intervalo
    expanded_df = pd.DataFrame(columns=['salario', 'Feminino', 'Masculino', 'min', 'max'])

    # Iterando sobre todos os salários e atribuindo os valores de "Feminino", "Masculino", "min" e "max"
    last_feminino = last_masculino = last_min = last_max = 0

    for s in salarios:
        # Encontrar a linha que corresponde ao intervalo em que o salário se encaixa
        for i, row in abs_genero.iterrows():
            if row['min'] <= s < row['max']:
                try:
                    last_feminino = row['Feminino']
                except:
                    last_feminino = 0
                last_masculino = row['Masculino']
                last_min = row['min']
                last_max = row['max']
                break
        
        # Adiciona a nova linha ao DataFrame expandido
        new_row = pd.DataFrame({
            'salario': [s],
            'Feminino': [last_feminino],
            'Masculino': [last_masculino],
            'min': [last_min],
            'max': [last_max]
        })
        
        # Concatena a nova linha ao DataFrame expandido
        expanded_df = pd.concat([expanded_df, new_row], ignore_index=True)
        # expanded_df.to_csv(f"profs/{prof}.csv")

```

```python
# Agrupa por profissão e gênero e calcula soma dos salários ponderados
grouped = df.groupby(["profissao", "genero"]).agg(
    total_salario=pd.NamedAgg(column="salario", aggfunc=lambda x: (x * 1).sum()),
    count=pd.NamedAgg(column="salario", aggfunc="count")
).reset_index()

# # Calcula a média salarial por grupo
grouped["media_salarial"] = grouped["total_salario"] / grouped["count"]

# Separa em dois DataFrames: feminino e masculino
feminino_df = grouped[grouped["genero"] == "Feminino"][["profissao", "media_salarial"]]
masculino_df = grouped[grouped["genero"] == "Masculino"][["profissao", "media_salarial"]]

# Renomeia colunas
feminino_df = feminino_df.rename(columns={"media_salarial": "media_feminina"})
masculino_df = masculino_df.rename(columns={"media_salarial": "media_masculina"})

# Junta os dois DataFrames
media_por_profissao = pd.merge(feminino_df, masculino_df, on="profissao", how="outer").fillna(0)

media_por_profissao.to_csv("profs/media.csv")
```

```python
import pandas as pd

# Suponha que df seja o seu DataFrame original

# 1. Filtra apenas gêneros de interesse
df = df[df[("('P1_b ', 'Genero')")].isin(["Feminino", "Masculino"])]

# 2. Corrige nomes das colunas para facilitar
df = df.rename(columns={
    "('P2_f ', 'Cargo Atual')": "cargo",
    "('P1_b ', 'Genero')": "genero",
})

# 3. Conta o número de mulheres e homens por cargo
gender_counts = df.pivot_table(index="cargo", columns="genero", values="salario", aggfunc="count", fill_value=0).reset_index()
gender_counts = gender_counts.rename(columns={"Feminino": "num_mulheres", "Masculino": "num_homens"})

# 4. Calcula salário médio, mínimo e máximo por cargo e gênero
salary_stats = df.groupby(["cargo", "genero"]).agg(
    salario_medio=('salario', 'mean'),
    salario_min=('salario', 'min'),
    salario_max=('salario', 'max')
).reset_index()

# 5. Junta as informações de contagem
final_df = salary_stats.merge(gender_counts, on="cargo", how="left")

# 6. Exporta para CSV
final_df.to_csv("salario_estatisticas_por_cargo_genero.csv", index=False)

print("Arquivo 'salario_estatisticas_por_cargo_genero.csv' criado com sucesso!")
```
