```python
import matplotlib.pyplot as plt

# Contagem de cada categoria de gênero
contagem_genero = df["('P1_b ', 'Genero')"].value_counts()

# Criando o gráfico de barras
contagem_genero.plot(kind='bar', color='skyblue', edgecolor='black')

# Títulos e rótulos
plt.title('Distribuição por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Frequência')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Renomear colunas 
col_genero = "('P1_b ', 'Genero')" 
col_area = "('P2_b ', 'Setor')"
col_cargo = "('P2_f ', 'Cargo Atual')"
col_salario = "('P2_h ', 'Faixa salarial')" #"('P2_o_a ', 'Remuneração/Salário')"

# Distribuição geral por gênero
genero_counts = df[col_genero].value_counts(normalize=True).sort_values(ascending=False)

plt.figure(figsize=(6, 4))
sns.barplot(x=genero_counts.index, y=genero_counts.values * 100, palette="pastel")
plt.ylabel('% do total')
plt.title('Distribuição por Gênero')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Proporção de gênero por área de atuação
area_genero = df.groupby([col_area, col_genero]).size().reset_index(name='contagem')
area_total = df.groupby(col_area).size().reset_index(name='total')
area_genero = area_genero.merge(area_total, on=col_area)
area_genero['percentual'] = 100 * area_genero['contagem'] / area_genero['total']

plt.figure(figsize=(12, 6))
sns.barplot(data=area_genero, x=col_area, y='percentual', hue=col_genero, palette="Set2")
plt.title('Proporção por Gênero em Cada Área de Atuação')
plt.ylabel('% dentro da área')
plt.xlabel('Área de Atuação')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gênero')
plt.tight_layout()
plt.show()
```

```python
# Dicionário de conversão de faixa salarial para valor médio
faixa_para_media = {
    'de R$ 1.000/mês a R$ 2.000/mês': 1500,
    'de R$ 2.001/mês a R$ 4.000/mês': 3000,
    'de R$ 4.001/mês a R$ 6.000/mês': 5000,
    'de R$ 6.001/mês a R$ 8.000/mês': 7000,
    'de R$ 8.001/mês a R$ 12.000/mês': 10000,
    'de R$ 12.001/mês a R$ 16.000/mês': 14000,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000,
    'de R$ 20.001/mês a R$ 25.000/mês': 22500,
    'acima de R$ 25.001/mês': 26000,
    'até R$ 1.000/mês': 500
}

# Aplicar mapeamento na coluna de salário
df["salario"] = df["('P2_h ', 'Faixa salarial')"].map(faixa_para_media)
```

```python
# Lista de fatores para analisar
fatores = ["('P1_e_b ', 'Regiao onde mora')", "('P2_i ', 'Quanto tempo de experiência na área de dados você tem?')", "('P2_f ', 'Cargo Atual')", "('P1_h ', 'Nivel de Ensino')"]

# Tamanho da figura
plt.figure(figsize=(16, 25))

# Gerar gráfico para cada fator
for i, fator in enumerate(fatores, 1):
    plt.subplot(len(fatores), 1, i)

    # Agrupar por fator e gênero e calcular média do salário numérico
    medias = df.groupby([fator, "('P1_b ', 'Genero')"])['salario'].mean().reset_index()

    # Ordenar valores do eixo x (opcional)
    ordenado = df[fator].value_counts().index

    # Gráfico de barras
    sns.barplot(data=medias, x=fator, y='salario', hue="('P1_b ', 'Genero')", order=ordenado)

    plt.title(f'Salário médio por Gênero e {fator}')
    plt.xlabel(fator)
    plt.ylabel('Salário médio (R$)')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
